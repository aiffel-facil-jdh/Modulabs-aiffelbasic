## 풀이용 소스

```python
def solution(N):
    #N = str(input())
    if len(N) < 2:
        N = '0' + N
    print(N[0], N[1])
    
    cnt = 1
    tmp_1, tmp_2 = N[0], N[-1]
    print(tmp_1, tmp_2)
    
    ans = ''
    while 1:
        sum = int(tmp_1) + int(tmp_2)
        print(sum)
        tmp_1 = tmp_2
        tmp_2 = str(sum)[-1]
        ans = str(tmp_1) + str(tmp_2)
        print(ans)
        
        if N == ans:
            print(f"Cicle: {cnt}")
            break
        else:
            cnt += 1
            
```

```python
solution('1')
```

## 백준 채점용 소스

```python
N = str(input())
if len(N) < 2:
    N = '0' + N

cnt = 1
tmp_1, tmp_2 = N[0], N[-1]
print(tmp_1, tmp_2)

ans = ''
while 1:
    sum = int(tmp_1) + int(tmp_2)
    tmp_1 = tmp_2
    tmp_2 = str(sum)[-1]
    ans = str(tmp_1) + str(tmp_2)

    if N == ans:
        print(cnt)
        break
    else:
        cnt += 1
```
